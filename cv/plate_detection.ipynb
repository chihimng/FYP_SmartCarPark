## Setup

```python
import matplotlib.pyplot as plt
import imutils
import cv2
import numpy as np
```

## Import image

```python
src = cv2.imread("./images/1.png")
```

```python
plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
_ = plt.xticks([]), plt.yticks([])
```

```python
bw = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
plt.imshow(bw, cmap = 'gray', )
_ = plt.xticks([]), plt.yticks([])
```

```python
bw_blur = cv2.blur(bw, (1,1))
plt.imshow(bw_blur, cmap = 'gray', )
_ = plt.xticks([]), plt.yticks([])
```

```python
threshold_min = 200
ratio_min_max = 1
bw_canny = cv2.Canny(bw_blur, threshold_min, threshold_min*ratio_min_max, apertureSize=5)
plt.imshow(bw_canny, cmap = 'gray', )
_ = plt.xticks([]), plt.yticks([])
```

```python
lines = cv2.HoughLinesP(bw_canny, 1, np.pi/180, 70, minLineLength=30, maxLineGap=10)[0]
len(lines)
```

```python
src_lines = src.copy()    
for x1,y1,x2,y2 in lines:
    cv2.line(src_lines,(x1,y1),(x2,y2),(0,255,0),2)
    
plt.imshow(cv2.cvtColor(src_lines, cv2.COLOR_BGR2RGB))
_ = plt.xticks([]), plt.yticks([])
```
